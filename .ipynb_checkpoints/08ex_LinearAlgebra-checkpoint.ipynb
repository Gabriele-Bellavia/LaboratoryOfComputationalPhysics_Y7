1\. **PCA on 3D dataset**

* Generate a dataset with 3 features each with N entries (N being ${\cal O}(1000)$). With $N(\mu,\sigma)$ the normali distribution with mean $\mu$ and $\sigma$  standard deviation, generate the 3 variables $x_{1,2,3}$ such that:
    * $x_1$ is distributed as $N(0,1)$
    * $x_2$ is distributed as $x_1+N(0,3)$
    * $x_3$ is given by $2x_1+x_2$
* Find the eigenvectors and eigenvalues of the covariance matrix of the dataset
* Find the eigenvectors and eigenvalues using SVD. Check that the two procedures yield to same result
* What percent of the total dataset's variability is explained by the principal components? Given how the dataset was constructed, do these make sense? Reduce the dimensionality of the system so that at least 99% of the total variability is retained.
* Redefine the data in the basis yielded by the PCA procedure
* Plot the data points in the original and the new coordiantes as a set of scatter plots. Your final figure should have 2 rows of 3 plots each, where the columns show the (0,1), (0,2) and (1,2) proejctions.


```python
import scipy as sp
from scipy import linalg as la
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
np.random.seed(589345)
N = 1001
x1 = np.random.normal(0,1, N)
x2 = x1 + np.random.normal(0,3,N)
x3 = 2 * x1 + x2
data = np.array([x1, x2, x3]).T
data = data - np.mean(data, axis=0, keepdims=True)
data = pd.DataFrame(data, columns=['x1', 'x2', 'x3'])

##Eigenvalues
covariance_matrix = np.cov(data.T)
#print(covariance_matrix)
l, v = la.eig(covariance_matrix)
l=l.real
print(f"Eigenvalues:\n{l},\nEigenvectors:\n{v}\n")

##SVD
U, S, VT = la.svd(data.T)

l_svd = (S ** 2) / (N-1)
v_svd = U

print(f"SVD results:\nVT:\n{U}\n\nl_svd:\n{l_svd}\n")
atol=0.1

if np.allclose(np.sort(l), np.sort(l_svd), atol=atol) == True and np.allclose(np.sort(v_svd), np.sort(v), atol=atol) == True:
    print("The two methods yield to the same result.")
else: print("The two methods don't yield to the same result.")

#percent of data
variability = np.round(l_svd / np.sum(l_svd) * 100, 2)

new_data = np.dot(v.T, data.T)
new_data = pd.DataFrame(new_data.T, columns=['x1','x2','x3'])
#new_data.head()
#fig, axes = plt.subplots(3,3,figsize=(10,10))
#for i in range(3):
 #   for j in range(3):
  #      axes[i,j].scatter(new_data[i],new_data[j])
#print(new_data)
scale_factor = 1
fig= plt.figure(figsize=(10,10))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(data['x1'], data['x2'], data['x3'], alpha=.01, color='blue')
ax.scatter(new_data['x1'], new_data['x2'], new_data['x3'], alpha=.01, color='red')

#plt.ylim(-0.1,0.1)
#scale_factor=1
ax.view_init(elev=40, azim=20)

for li, vi in zip(l, v):
    ax.plot(
        [0, scale_factor * li * vi[0]],  # x = asse x3 (per come li hai scambiati)
        [0, scale_factor * li * vi[1]],  # y = asse x2
        [0, scale_factor * li * vi[2]],  # z = asse x1
        color='green', lw=3
    )
plt.show()
```
